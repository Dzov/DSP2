# Data pre-analysis

## Null data

Identifying amount of null data by column

```python
print(raw_data.isnull().mean())

print(raw_data.isnull().mean().mean())
```

Removing columns with no data at all

```python
null_column_data = raw_data.isna().mean()
columns_to_remove = null_column_data[null_column_data == 1].index
print("There is " + str(columns_to_remove.size) + " column without data to remove")

data = raw_data.drop(columns_to_remove, axis=1)
data.shape
```

Removing rows with no data at all throughout the years

```python
years = data.columns[4:].values

null_row_data = raw_data[years].isnull().mean(axis=1)
rows_to_remove = null_row_data[null_row_data == 1].index
print(
    "There are "
    + str(rows_to_remove.size)
    + " rows that only contain null values. Which is "
    + str(rows_to_remove.size / raw_data.shape[0] * 100)
    + "% of the total "
    + str(raw_data.shape[0])
    + " rows"
)

data = data.drop(rows_to_remove, axis=0)
data.shape
```

Identifying the years with the most data

```python
plt.figure(figsize=[15, 6])
sns.barplot(x=years, y=data.iloc[:, 4:].notna().mean() * 100)
plt.xticks(rotation=90)
plt.grid(axis="y")
plt.title("Percentage of non null values by year")
plt.show()
```

This graph shows that there are several years with only 10% or less exploitable data. We will focus on the years with more than 20% available data, along with future years to allow for projection. 

```python
years_to_remove = [
    "1971",
    "1972",
    "1973",
    "1974",
    "1976",
    "1977",
    "1978",
    "1979",
    "1981",
    "1982",
    "1983",
    "1984",
    "1986",
    "1987",
    "1988",
    "1989",
    "2016",
    "2017",
]

filtered_years = np.setdiff1d(years, years_to_remove)
data = data.drop(columns=years_to_remove)
```

```python
def remove_empty_rows_and_columns(df: pd.DataFrame, subset: list = None):
    df = df.dropna(axis=1, how="all")

    subset = subset if subset is not None else df.columns[4:-1]
    df = df.dropna(axis=0, how="all", subset=subset)

    return df


def show_missing_values(df, title: str):
    missing_values = df.isna().mean() * 100
    plt.figure(figsize=(12, 3))
    sns.lineplot(data=missing_values[4:-1])
    plt.xticks(rotation=90)
    plt.xlabel("Year")
    plt.ylabel("Percentage of missing values")
    plt.title(title)
    plt.show()
    return missing_values
```
