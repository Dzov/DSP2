# Analysing chosen indicators

Defining methods to explore and clean the data sets

```python
def getDataByIndicatorNames(indicators: list):
    df = data[data["Indicator Name"].isin(indicators)]

    return df


def removeNullValues(df: pd.DataFrame, subset: list = None):
    df = df.dropna(axis=1, how="all")
    subset = subset if subset is not None else df.columns[4:-1]

    return df.dropna(axis=0, how="all", subset=subset)


def showMissingValues(df, title: str):
    missingValues = df.isna().mean() * 100
    plt.figure(figsize=(12, 3))
    sns.lineplot(data=missingValues[4:-1])
    plt.xticks(rotation=90)
    plt.xlabel("Year")
    plt.ylabel("Percentage of missing values")
    plt.title(title)
    plt.show()
    return missingValues


def removeInvalidCountries(df: pd.DataFrame):
    indexedCountries = country.set_index("Country Code")
    invalidCountries = indexedCountries.loc[indexedCountries["Region"].isna()].index

    return df[~df["Country Code"].isin(invalidCountries)]


def groupByCountries(df: pd.DataFrame):
    return (
        df.groupby("Country Name", axis="rows")
        .sum(numeric_only=True)
        .sort_values(df.columns[-1], ascending=False)
    )
```

## Education

### 15-24 Population Attendance Rate

- Barro-Lee: Percentage of population age 15-19 with secondary schooling. Total (Incomplete and Completed Secondary)
- Barro-Lee: Percentage of population age 15-19 with tertiary schooling. Total (Incomplete and Completed Tertiary)

```python
education1519 = getDataByIndicatorNames(
    [
        "Barro-Lee: Percentage of population age 15-19 with secondary schooling. Total (Incomplete and Completed Secondary)",
        "Barro-Lee: Percentage of population age 15-19 with tertiary schooling. Total (Incomplete and Completed Tertiary)",
    ]
)

showMissingValues(
    education1519,
    "Missing values by year for 15-19yo indicators",
)

education1519 = groupByCountries(removeNullValues(education1519))

education1519["mean"] = education1519.mean(axis=1)
education1519["std"] = education1519.std(axis=1)
education1519["median"] = education1519.median(axis=1)

display(education1519.head(10))
```

- Barro-Lee: Percentage of population age 20-24 with secondary schooling. Total (Incomplete and Completed Secondary)
- Barro-Lee: Percentage of population age 20-24 with tertiary schooling. Total (Incomplete and Completed Tertiary)

```python
education2024 = getDataByIndicatorNames(
    [
        "Barro-Lee: Percentage of population age 20-24 with secondary schooling. Total (Incomplete and Completed Secondary)",
        "Barro-Lee: Percentage of population age 20-24 with tertiary schooling. Total (Incomplete and Completed Tertiary)",
    ]
)

e = showMissingValues(
    education2024, "Missing values by year for 20-24yo enrolment indicators"
)

display(e.mean())

education2024 = groupByCountries(removeNullValues(education2024))
education2024["mean"] = education2024.mean(
    axis=1,
)
education2024["std"] = education2024.std(axis=1)
education2024["median"] = education2024.median(axis=1)
education2024 = education2024.sort_values("mean", ascending=False)

topHalfCountries = education2024.iloc[: (len(education2024) // 2)]
bottomHalfCountries = education2024.iloc[(len(education2024) // 2) :]

fig, ax = plt.subplots(figsize=(12, 3))
sns.lineplot(y=topHalfCountries["mean"], x=topHalfCountries.index, label="Mean")
sns.lineplot(y=topHalfCountries["median"], x=topHalfCountries.index, label="Median")
sns.lineplot(y=topHalfCountries["std"], x=topHalfCountries.index, label="Std")
plt.xticks(rotation=90)

plt.title("Mean, Median, and Standard Deviation by Country")
plt.xlabel("Top Half Countries by mean")
plt.ylabel("Value")

ax.legend()
ax.grid(True)
plt.show()

fig, ax = plt.subplots(figsize=(12, 3))
sns.lineplot(y=bottomHalfCountries["mean"], x=bottomHalfCountries.index, label="Mean")
sns.lineplot(
    y=bottomHalfCountries["median"], x=bottomHalfCountries.index, label="Median"
)
sns.lineplot(y=bottomHalfCountries["std"], x=bottomHalfCountries.index, label="Std")
plt.xticks(rotation=90)

plt.title("Mean, Median, and Standard Deviation by Country")
plt.xlabel("Bottom Half Countries by mean")
plt.ylabel("Value")

ax.legend()
ax.grid(True)
plt.show()
```

# Scoring 

Youth enrolment statistics 

```python
youthEnrolment2010 = removeNullValues(
    youthPopulation["2010"]
    .multiply(education2024["2010"] / 100)
    .to_frame()
    .rename(columns={"2010": "Youth Enrolment"}),
    ["Youth Enrolment"],
)

youthEnrolment2010["score"] = (
    youthEnrolment2010["Youth Enrolment"] / youthEnrolment2010["Youth Enrolment"].max()
)
display(youthEnrolment2010.sort_values("score", ascending=False).head())
```

Internet statistics 

```python
internetUsers2015 = (
    youthPopulation["2015"]
    .multiply(internetUsers["2015"] / 100)
    .to_frame()
    .rename(columns={"2015": "Internet Users"})
)

internetUsersStats = internetUsers["2015"] / internetUsers["2015"].max()
internetUsersStats = internetUsersStats.to_frame().rename(
    columns={"2015": "Percentage Score"}
)
internetUsersStats["Total Score"] = (
    internetUsers2015["Internet Users"] / internetUsers2015["Internet Users"].max()
)
display(internetUsersStats.sort_values("Percentage Score", ascending=False).head())
```

Summary 

```python
statistics = internetUsersStats
statistics["Enrolment Score"] = youthEnrolment2010["score"]
statistics["Mean"] = statistics.mean(axis=1, skipna=False)

statistics = statistics.dropna(axis=0).sort_values("Mean", ascending=False)
display(statistics.head())

statisticsByRegion = (
    statistics.merge(
        country[["Short Name", "Region"]], left_on="Country Name", right_on="Short Name"
    )
    .groupby("Region")
    .mean(numeric_only=True)
    .sort_values("Mean", ascending=False)
)

display(statisticsByRegion)
statisticsByRegion["Mean"].plot(kind="bar")
plt.xticks(rotation=45)
plt.ylabel("Score")
plt.title("Target regions based on enrolment rate, population and internet access")
```
