**Identifying and filling in the missing data**

Zoom in on sample data

```python
def merge_regions(df):
    return df.merge(
        countries[["Short Name", "Region"]],
        left_on="Country Name",
        right_on="Short Name",
    ).rename(columns={"Short Name": "Country Name"})


education_enrolment = merge_regions(education_enrolment_by_filtered_country)


def get_sample_countries_by_region(df, number=2):
    sample_countries = []
    for region in regions:
        sample_countries.append(
            df.loc[df["Region"] == region, "Country Name"].sample(n=number)
        )
    return df.loc[
        df["Country Name"].isin(
            [
                item
                for sample_region_countries in sample_countries
                for item in sample_region_countries
            ]
        )
    ]
```

Preview a sample of countries with missing data

```python
education_countries_missing_data = education_enrolment.loc[
    education_enrolment.isna().mean(axis=1) > 0
]

sample_with_missing_data = get_sample_countries_by_region(
    education_countries_missing_data, 1
)


def melt_df(df):
    return pd.melt(
        df,
        id_vars=["Region", "Country Name"],
        var_name="Year",
        value_name="Youth Population",
    )


def show_scatterplot(df):
    plt.figure(figsize=[10, 5])
    sns.scatterplot(
        x="Year",
        y="Youth Population",
        hue="Country Name",
        data=melt_df(df),
    )
    plt.legend(bbox_to_anchor=(1, 1))
    plt.show()


show_scatterplot(sample_with_missing_data)
```

Fill in missing values with different methods

```python
def fill_years(df, year_range, method):
    subset = np.intersect1d([str(item) for item in list(year_range)], df.columns)
    df[subset] = df[subset].fillna(method=method, axis=1)
    return df


dfs = [
    melt_df(fill_years(sample_with_missing_data.copy(), range(1999, 2016), "bfill")),
    melt_df(fill_years(sample_with_missing_data.copy(), range(1999, 2016), "ffill")),
    melt_df(sample_with_missing_data.copy().interpolate()),
]

titles = ["Bfill", "Ffill", "Interpolation"]

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

for i, df in enumerate(dfs):
    ax = axes[i]
    sns.scatterplot(
        x="Year",
        y="Youth Population",
        hue="Country Name",
        data=df,
        ax=ax,
    )
    ax.set_xticks(ax.get_xticks())
    ax.set_xticklabels(ax.get_xticklabels(), rotation=65)
    ax.set_title(titles[i])


plt.tight_layout()
plt.show()
```

The interpolation method yields values that are much too far from the previous and next values. 
The bfill and ffill methods yield good results but fail to fill all values, therefore a combination of these two methods with be used to fill the missing values of the entire data set. 

```python
def fill_missing_values(df, year_range):
    return fill_years(fill_years(df, year_range, "bfill"), year_range, "ffill")


education_enrolment = fill_missing_values(education_enrolment, range(1999, 2016))
education_enrolment = education_enrolment.set_index("Country Name")
education_enrolment.loc[education_enrolment.isna().mean(axis=1) > 0]
```
