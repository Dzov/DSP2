```python
def merge_regions(df):
    return df.merge(
        countries[["Short Name", "Region"]],
        left_on="Country Name",
        right_on="Short Name",
    ).rename(columns={"Short Name": "Country Name"})


education_enrolment = merge_regions(education_enrolment_by_filtered_country)


def get_sample_countries_by_region(df, number=1):
    sample_countries = []
    for region in regions:
        sample_countries.append(
            df.loc[df["Region"] == region, "Country Name"].sample(n=number)
        )
    return df.loc[
        df["Country Name"].isin(
            [
                item
                for sample_region_countries in sample_countries
                for item in sample_region_countries
            ]
        )
    ]
```

##### Filling in missing data

Fill in missing values with different methods

```python
def get_range_subset(df, year_range):
    return np.intersect1d([str(item) for item in list(year_range)], df.columns)


def fill_years(df, year_range, method):
    subset = get_range_subset(df, year_range)
    df[subset] = df[subset].fillna(method=method, axis=1)
    return df


dfs = [
    melt_df(
        fill_years(sample_with_missing_data.copy(), range(1999, 2016), "bfill"),
        "Number of Enroled Students",
    ),
    melt_df(
        fill_years(sample_with_missing_data.copy(), range(1999, 2016), "ffill"),
        "Number of Enroled Students",
    ),
    melt_df(
        sample_with_missing_data.copy().interpolate(), "Number of Enroled Students"
    ),
]

titles = ["Bfill", "Ffill", "Interpolation"]

fig, axes = plt.subplots(3, 1, figsize=(8, 10))

for i, df in enumerate(dfs):
    ax = axes[i]
    sns.scatterplot(
        x="Year",
        y="Number of Enroled Students",
        hue="Country Name",
        data=df,
        ax=ax,
    )
    ax.set_xticks(ax.get_xticks())
    ax.set_xticklabels(ax.get_xticklabels(), rotation=65)
    ax.set_title(titles[i])
    ax.legend(bbox_to_anchor=(1, 1))


plt.tight_layout()
plt.show()
```

=> The interpolation method yields values that are much too far from the previous and next values. 

=> The bfill and ffill methods yield good results but fail to fill all values, therefore a combination of these two methods with be used to fill the missing values of the entire data set. 

```python
def fill_missing_values(df, year_range):
    return fill_years(fill_years(df, year_range, "bfill"), year_range, "ffill")


clean_education_enrolment = fill_missing_values(education_enrolment, range(1999, 2016))
clean_education_enrolment.loc[clean_education_enrolment.isna().mean(axis=1) > 0]
```

##### Spotting potential outliers

```python
def show_outliers(df, value_name, title):
    melted_df = melt_df(df, value_name)

    plt.figure(figsize=[10, 30])
    sns.boxplot(data=melted_df, x=value_name, y="Country Name")
    plt.title(title)
    plt.show()


show_outliers(
    education_enrolment,
    "Number of Enroled Students",
    "Ditribution of Number of Enroled Students for Each Country",
)
```

```python
def remove_outliers(row):
    Q1 = np.percentile(row, 25)
    Q3 = np.percentile(row, 75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = {
        row.name: [
            (column_name, value)
            for value, column_name in zip(row, row.index)
            if value < lower_bound or value > upper_bound
        ]
    }

    if len(outliers[row.name]) > 0:
        print(outliers)

    return row[(row >= lower_bound) & (row <= upper_bound)]


clean_education_enrolment = education_enrolment.copy().set_index("Country Name")
clean_education_enrolment[
    get_range_subset(clean_education_enrolment, range(1999, 2016))
] = clean_education_enrolment[
    get_range_subset(clean_education_enrolment, range(1999, 2016))
].apply(
    remove_outliers, axis=1
)
```

Internet statistics
